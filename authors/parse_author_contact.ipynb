# Parse the author table
This will be used to format affiliations and funding.

```python
import pandas as pd
import numpy as np
import os, re
from io import StringIO
```

Convert the table into a simpler format and load into pandas.  Create author initials for funding.

```python
def initials(full_name):
    name_tokens = full_name.strip().split()
    name_tokens = [name[0] + '.' for name in name_tokens]
    return ''.join(name_tokens)
```

```python
with open(os.path.join('..','author_contact_approve.md'), 'r') as in_f:
    orig_file = in_f.read()

# prepare for pandas
mod_file = orig_file.replace('|---------------|-----------|--------|-----------------|-------|---------|----------------|\n','')
mod_file = mod_file.replace('  ',' ')
mod_file = mod_file.replace(' | ','\t')
mod_file = mod_file.replace('| ','')
mod_file = mod_file.replace(' |','')
mod_file = mod_file.replace('|','')
# temporary until I look into character encoding problems
mod_file = mod_file.replace('Ã¤','ä')
mod_file = mod_file.replace('Ã¼','ü')


author_data = StringIO(mod_file)

author_df = pd.read_csv(author_data, sep='\t')

# add initials
author_df['Initials'] = author_df['Full name'].apply(initials)

author_df.head(50)
```

Check for approval and unique initials

```python
assert len(author_df['Initials'].unique()) == len(author_df)
assert np.all(author_df['Approve'] == 'Yes')
```
